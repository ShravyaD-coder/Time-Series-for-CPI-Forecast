```python
import warnings

import matplotlib as mpl
import numpy as np
import pandas as pd  # Basic library for all of our dataset operations
import statsmodels as sm
from matplotlib import pyplot as plt
from pylab import rcParams
from sklearn.linear_model import LinearRegression
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa import api as smt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
```

```python
data = pd.read_csv("./Urban Actuals.csv")
data=data.dropna()
data
```

```python
data.columns=data.columns.str.strip()
data.columns=data.columns.str.replace(" ","_")
data.columns
```

```python
data['Date']=pd.to_datetime(data['Date'],dayfirst=True)
data=data.set_index('Date')
data.head()
```

```python
data_filtered=pd.DataFrame(data['Milk_and_products'])
data_filtered
```

```python
plt.plot(data_filtered,label='Cereals')
plt.title("Urban CPI of Cerelas")
plt.legend(loc="upper left")
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose


rcParams['figure.figsize'] = 18, 8
plt.figure(num=None, figsize=(50, 20), dpi=80, facecolor='w', edgecolor='k')
series = data_filtered
result = seasonal_decompose(series, model='mutiplicative',period=12)
result.plot();
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(data_filtered, lags=30); #q=10
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_pacf(data_filtered, lags=30, method='ywm'); #p=2
```

```python
from statsmodels.tsa.stattools import adfuller

result=adfuller(data_filtered)
print("p-value: ",result[1])

result=adfuller(data_filtered.diff().dropna())
print("p-value: ",result[1])

result=adfuller(data_filtered.diff().diff().dropna())
print("p-value: ",result[1])
```

```python
#train-test split


forecast_horizon=6
train=data_filtered[:'2022-09-01']
test=data_filtered['2022-10-01':]

print(train)
print(test)
#print(train.shape())
#print(test.shape())
```

```python
from statsmodels.tsa.arima.model import ARIMA
arima_model=ARIMA(train,order=(3,1,11))
model=arima_model.fit()
print(model.summary())
```

```python
y_pred=model.forecast(6)
y_pred
```

```python
data_copy=data_filtered.copy()
data_copy['ARIMA Manual']=[None]*len(train)+list(y_pred)
val=['Milk_and_products','ARIMA Manual']
data_copy[val].plot()
```
